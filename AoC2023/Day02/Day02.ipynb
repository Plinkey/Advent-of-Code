# Process Data

```python
def processResult(result):
    answer = {}
    for color in result:
        numberstr, color = color.split(' ')
        number = int(numberstr)
        answer[color] = number
    return answer
```

```python
def processResultLine(line):
    answer = []
    for result in line:
        answer.append(processResult(result))
    return answer
```

```python
gameno = []
gameresults = []
for line in rawdata:
    
    # Find Gamenumber
    gamestr, resultstr  = line.split(':')
    gameno.append(int(gamestr.split(' ')[-1]))
    
    # Process Game Results
    resultstr = resultstr.lstrip(' ')
    a = [i.strip(' ') for i in resultstr.split(';')]
    
    result = []
    for line in a:
        result.append([ii.strip(' ') for ii in line.split(',')])

    gameresults.append(processResultLine(result))
    
    
data = dict(zip(gameno, gameresults))
```

# Part One

```python
max_red = 12
max_green = 13
max_blue = 14
```

```python
badgames = []
for gameno, game in data.items():
    badgame = ''
    for trial in game:
        for color, number in trial.items():
            if color == 'red':
                if number > max_red:
                    badgame = gameno
            elif color == 'green':
                if number > max_green:
                    badgame = gameno
            elif color == 'blue':
                if number > max_blue:
                    badgame = gameno
    if badgame != '':
        badgames.append(badgame)


print('The answer to part one is: {}'.format(sum(data.keys())-sum(badgames)))
```

```python
def partTwo_processgame(game):
    maxRed = 0
    maxGreen = 0
    maxBlue = 0
    for trial in game:
        for color, number in trial.items():
            if color == 'red':
                if number > maxRed:
                    maxRed = number
            elif color == 'green':
                if number > maxGreen:
                    maxGreen = number
            elif color == 'blue':
                if number > maxBlue:
                    maxBlue = number
    return maxRed * maxBlue * maxGreen
```

```python
powers = []
for gameno, game in data.items():
    powers.append(partTwo_processgame(game))

print('The answer to part two is: {}'.format(sum(powers)))
```
