# Load data.  Convert str to int.  Keep '' on empty lines.

```python
#with open('Day01.example','r') as f:
with open('Day01.input','r') as f:
    rawdata = f.read().splitlines()
data = []
for line in rawdata:
    try:
        data.append(int(line))
    except:
        data.append('')
```

Loop through data, caluclate sum for each elf

```python
elf_total = [0]
for item in data:
    if item == '':
        elf_total.append(0) # add new entry for next elf
        continue
    else:
        elf_total[-1] += item
```

# PART ONE:  Maximum Calories

```python
print(" YOUR ANSWER TO PART IS: {}".format(max(elf_total)))
```

# PART TWO: TOTAL CALORIES OF TOP THREE

```python
total = 0
counter = 0
while counter < 3:
    total += elf_total.pop(elf_total.index(max(elf_total)))
    counter += 1
print(" YOUR ANSER TO PART TWO IS: {}".format(total))
```
