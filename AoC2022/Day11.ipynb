```python
example = False
if example:
    inputFile = 'Day11.example'
else:
    inputFile = 'Day11.input'

with open(inputFile, 'r') as f:
    rawdata = f.read().splitlines()
counter = 0
data = []
while counter < len(rawdata)-5:
    data.append(rawdata[counter:counter+6])
    counter += 7
```

```python
class Monkey():
    def __init__(self, varsIn):
        self.monkeyNum = varsIn[0]
        self.curItems = varsIn[1]
        self.operation = varsIn[2]
        self.divTest = varsIn[3]
        self.testTrue = varsIn[4]
        self.testFalse = varsIn[5]
        self.inspectCount = 0

    def processNextItem(self):
        self.inspectCount += 1
        item = self.curItems.pop(0)
        if self.operation[1] == '+':
            if self.operation[2] == 'old':
                item = item + item
            else:
                item = item + self.operation[2]
        else:
            if self.operation[2] == 'old':
                item = item * item
            else:
                item = item * self.operation[2]
        
        # Monkey Bored
        item = item // 3

        #Test:
        if item%self.divTest == 0:
            print('Moving {} to monkey {}'.format(item, self.testTrue))
            animals[self.testTrue].curItems.append(item)
        else:
            print('Moving {} to monkey {}'.format(item, self.testFalse))
            animals[self.testFalse].curItems.append(item)
        
    def runTurn(self):
        while len(self.curItems) > 0:
            self.processNextItem()
```

```python
def parseMonkey(stringList):
    monkeyNum = stringList[0].split(' ')[1].rstrip(':')
    monkeyNum = int(monkeyNum)
    # Starting Items
    tmp = stringList[1].split(':')[1].strip(' ')
    tmp = tmp.split(',')
    items = [int(i) for i in tmp]
    
    #Operation
    tmp = stringList[2].split('=')[-1].strip(' ')
    tmp = tmp.split(' ')
    oper = []
    for i in tmp:
        try:
            oper.append(int(i))
        except:
            oper.append(i)

    # Test
    test = int(stringList[3].split(' ')[-1])

    # If True
    ifTrue = int(stringList[4].split(' ')[-1])

    # If False
    ifFalse = int(stringList[5].split(' ')[-1])

    return [monkeyNum, items, oper, test, ifTrue, ifFalse]
```

```python
animals = {}
for m in data:
    a = parseMonkey(m)
    animals[a[0]] = Monkey(a)
```

```python
animals = {}
for m in data:
    a = parseMonkey(m)
    animals[a[0]] = Monkey(a)
counter = 1

while counter <= 20:
    for a in animals:
        animals[a].runTurn()
    counter += 1
    print("Counter now at: {}".format(counter))
```
