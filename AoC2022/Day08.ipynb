```python
import numpy as np
example = False
if example:
    inputFile = 'Day08.example'
else:
    inputFile = 'Day08.input'

with open(inputFile, 'r') as f:
    rawdata = f.read().splitlines()
data = []
for line in rawdata:
    row = []
    for char in line:
        row.append(int(char))
    data.append(row)
```

Create mask.  1 = visible from outside, 0 = not visible (or unknown)

```python
mask = []
for line in data:
    row = []
    for tree in line:
        row.append(0)
    mask.append(row)

for idx,_ in enumerate(mask):
    for jdx,_ in enumerate(mask[idx]):
        if idx == 0:
            mask[idx][jdx] = 1
        elif idx == len(mask)-1:
            mask[idx][jdx] = 1
        elif jdx == 0:
            mask[idx][jdx] = 1
        elif jdx == len(mask[0])-1:
            mask[idx][jdx] = 1

```

```python
def checkNorth(row, column):
    curValue = data[row][column]
    others = []
    counter = row - 1
    while counter >= 0:
        others.append(data[counter][column])
        counter -= 1
    if all(x < curValue for x in others):
        return True
    else:
        return False
```

```python
def checkWest(row, column):
    curValue = data[row][column]
    others = []
    counter = column - 1
    while counter >= 0:
        others.append(data[row][counter])
        counter -= 1
    if all(x < curValue for x in others):
        return True
    else:
        return False
```

```python
def checkEast(row, column):
    curValue = data[row][column]
    others = []
    counter = column + 1
    while counter < len(data[row]):
        others.append(data[row][counter])
        counter += 1
    if all(x < curValue for x in others):
        return True
    else:
        return False
```

```python
def checkSouth(row,column):
    curValue = data[row][column]
    others = []
    counter = row + 1
    while counter < len(data):
        others.append(data[counter][column])
        counter += 1
    if all(x < curValue for x in others):
        return True
    else:
        return False
```

```python
def checkAll(row, column):
    if checkNorth(row, column):
        if checkWest(row, column):
            if checkEast(row, column):
                if checkSouth(row,column):
                    return True
    else:
        return False
```

```python
def checkAny(row, column):
    if checkNorth(row, column):
        return True
    elif checkWest(row,column):
        return True
    elif checkEast(row, column):
        return True
    elif checkSouth(row, column):
        return True
    else:
        return False
```

```python
def countVis():
    count = 0
    for idx,_ in enumerate(mask):
        for jdx,_ in enumerate(mask[idx]):
            if mask[idx][jdx] == 1:
                count += 1
    return count
```

```python
for idx,_ in enumerate(data):
    for jdx,_ in enumerate(data[idx]):
        if idx == 0:
            continue
        elif idx == len(data)-1:
            continue
        elif jdx == 0:
            continue
        elif jdx == len(data[idx]) - 1:
            continue
        elif checkAny(idx, jdx):
            mask[idx][jdx] = 1

print("The answer to part one is: {}".format(countVis()))
```

```python
def checkOthers(curValue, others):
    total = 0
    for val in others:
        if curValue > val:
            total += 1
            continue
        else:
            total += 1
            return total
    return total


def visNorth(row, column):
    curValue = data[row][column]
    others = []
    counter = row - 1
    while counter >= 0:
        others.append(data[counter][column])
        counter -= 1
    return checkOthers(curValue, others)



def visWest(row, column):
    curValue = data[row][column]
    others = []
    counter = column - 1
    while counter >= 0:
        others.append(data[row][counter])
        counter -= 1
    return checkOthers(curValue, others)


def visEast(row, column):
    curValue = data[row][column]
    others = []
    counter = column + 1
    while counter < len(data[row]):
        others.append(data[row][counter])
        counter += 1
    return checkOthers(curValue, others)


def visSouth(row, column):
    curValue = data[row][column]
    others = []
    counter = row + 1
    while counter < len(data):
        others.append(data[counter][column])
        counter += 1
    return checkOthers(curValue, others)
```

```python
def visScore(row, column):
    a = visNorth(row, column)
    b = visWest(row, column)
    c = visEast(row, column)
    d = visSouth(row, column)
    return a*b*c*d
```

```python
maxScore = 0
for idx,_ in enumerate(data):
    for jdx,_ in enumerate(data[idx]):
        curScore = visScore(idx, jdx)
        if curScore > maxScore:
            maxScore = curScore

print('The answer to part two is: {}'.format(maxScore))
```
