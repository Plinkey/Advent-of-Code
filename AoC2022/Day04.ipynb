```python
#with open('Day04.example','r') as f:
with open('Day04.input','r') as f:
    rawdata = f.read().splitlines()
data = []
for l in rawdata:
    tmp = (l.split(','))
    a = tmp[0].split('-')
    first = int(a[0])
    second = int(a[1])
    
    b = tmp[1].split('-')
    third = int(b[0])
    fourth = int(b[1])

    data.append([[first, second], [third, fourth]])
```

```python
data
```

# PART ONE

```python
count = 0
for line in data:
    lrange = [i for i in range(line[0][0],line[0][1]+1)]
    rrange = [i for i in range(line[1][0], line[1][1]+1)]
    if (all(x in rrange for x in lrange)):
        count += 1
        continue
    elif (all(x in lrange for x in rrange)):
        count += 1
        continue

print("Answer to part one is: {}".format(count))
```

# PART TWO

```python
count = 0
for line in data:
    lrange = [i for i in range(line[0][0],line[0][1]+1)]
    rrange = [i for i in range(line[1][0], line[1][1]+1)]
    if (any(x in rrange for x in lrange)):
        count += 1
        continue
    elif (any(x in lrange for x in rrange)):
        count += 1
        continue

print("Answer to part two is: {}".format(count))
```
