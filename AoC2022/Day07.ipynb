```python
example = False
if example:
    inputFile = 'Day07.example'
else:
    inputFile = 'Day07.input'
with open(inputFile,'r') as f:
    rawdata = f.read().splitlines()
data = rawdata
```

```python
def findCDCommand(dirname):
    searchStr = '$ cd '+dirname
    idxs = []
    for idx, val in enumerate(data):
        if val == searchStr:
            idxs.append(idx)
    return idxs
```

```python
def dirContents(dirname):
    idxs = findCDCommand(dirname)
    contents = []
    for idx in idxs:
        if data[idx+1] == '$ ls':
            contentIdx = idx+2
            while True:
                if '$ ' in data[contentIdx]:
                    break
                else:
                    contents.append(data[contentIdx])
                    contentIdx += 1
                    if contentIdx >= len(data):
                        break
        else:
            continue
    return contents
```

```python
class directory:
    def __init__(self, contents):
        self.subdirs = []
        self.files = []
        for line in contents:
            if 'dir ' in line:
                self.subdirs.append(line.split(' ')[1])
            else:
                self.files.append(line)
        self.completeFlag = False
        self.checkComplete()
    def calcSize(self):
        sum = 0
        for item in self.files:
            sum += int(item.split(' ')[0])
        return sum
    def checkComplete(self):
        if self.subdirs == []:
            self.completeFlag = True
        return self.completeFlag
    
```

```python
def findAllDirs():
    l = []
    for line in data:
        if 'dir ' in line:
            if line not in l:
                l.append(line.split(' ')[1])
    return l
```

Loop through directories, filling in completed directories

```python
def loop(dirs):
    for d in dirs:
        curDir = dirs[d]
        if curDir.completeFlag == True:
            continue
        else:
            for i in curDir.subdirs:
                if dirs[i].completeFlag == True:
                    for f in dirs[i].files:
                        curDir.files.append(f)
                    curDir.subdirs.pop(curDir.subdirs.index(i))
                curDir.checkComplete()
```

Check if we're done looping

```python
def doneLoops(dirs):
    for d in dirs:
        if dirs[d].completeFlag == False:
            return False
    return True
```
