```python
import numpy as np

#with open('Day13.example','r') as f:
with open('Day13.input','r') as f:
    rawdata = f.read().splitlines()

points = []
folds = []
foldFlag = False
for line in rawdata:
    if foldFlag:
        tmp = line.split()[2]
        tmp = tmp.split('=')
        tmp[1] = int(tmp[1])
        folds.append(tmp)
        continue
    elif line == '':
        foldFlag = True
        continue
    else:
        tmp = [int(i) for i in line.split(',')]
        tmp2 = [tmp[1], tmp[0]]  #Flip so vertical is first number
        points.append(tmp2)
print(points)
print(folds)
```

# Find Size of Field

```python
nVert = 0
nHoriz = 0
for point in points:
    if point[0] > nVert:
        nVert = point[0]
    if point[1] > nHoriz:
        nHoriz = point[1]

nVert += 1
nHoriz += 1
        
print(nVert)
print(nHoriz)
```

# Seed Field with Zeros

```python
field = np.zeros([nVert, nHoriz])
```

# Draw Initial Points

```python
def draw(p):
    field[p[0],p[1]] = 1
```

```python
for p in points:
    draw(p)
print(field)
```

# Fold

```python
def folded_points(fold):
    # Return location of points BEYOND the fold
    
    if fold[0] == 'y': # Vertical fold
        folded = field[fold[1]-1:,:]
        tmp = np.where(folded==1)
        corrected = tmp[0] + fold[1]-1
        return list(zip(corrected,tmp[1]))
    
    if fold[0] == 'x': #Horizontal fold
        folded = field[:,fold[1]-1:]
        tmp = np.where(folded==1)
        corrected = tmp[1] + fold[1]-1
        return list(zip(tmp[0],corrected))
    
```

```python
def transpose(points, fold):
    if fold[0] == 'y': #Vertical fold
        transposed = []
        for p in points:
            newV = p[0] - 2*(p[0]-fold[1])
            transposed.append([newV, p[1]])
    if fold[0] == 'x': #Horizontal fold
        transposed = []
        for p in points:
            newH = p[1] - 2*(p[1]-fold[1])
            transposed.append([p[0], newH])
    return transposed
```

```python
def new_field(points, fold):
    if fold[0] == 'y': # Vertical fold
        newVert = fold[1]
        newHoriz = nHoriz
        for p in points:
            draw(p)
        return field[0:newVert,:]
    if fold[0] == 'x': #Horizontal fold
        newVert = nVert
        newHoriz = fold[1]
        for p in points:
            draw(p)
        return field[:,0:newHoriz]
```

# RUN PART ONE

```python
f =folded_points(folds[0])
t = transpose(f,folds[0])
field = new_field(t, folds[0])
```

```python
print('Answer to part one is: {}'.format(np.count_nonzero(field==1)))
```

```python
def printit(field):
    for row in field:
        p = []
        for x in row:
            if x == 0:
                p.append('.')
            if x == 1:
                p.append('#')
        print(''.join(p))
```
