# Part 1
## Return column as list

```
def column(col, data):
    ans = []
    for l in data:
        for idx, value in enumerate(l):
            if idx == col:
                ans.append(value)
    return ans
```

## Return Most Frequent

```
def most_freq(line):
    return mode(line)[0][0]
```

## Loop through input, return gamma (also epsilon)

```
def gamma(data):
    gamma = []
    for idx in range(len(data[0])):  # for number of columns
        gamma.append(most_freq(column(idx,data)))
    gamma_str = [str(q) for q in gamma]
    gamma = ''.join(gamma_str)
    epsilon = ''.join('1' if x == '0' else '0' for x in gamma_str)
    return int(gamma,2), int(epsilon,2)
```

## Run it

```
g, e = gamma(data)
print('Answer part 1: {}'.format(g*e))
```

## Filter by Most Common in bit

```
def most_filter(col, data):
    def freq(line):
        value, freq = mode(line)
        if freq == len(line)/2: # if equal number of bits with both 1's and 0's
            return 1
        else:
            return value[0]
    '''Keep only the lines in data that have the most common value in column "col"'''
    c = column(col, data)
    key = freq(c)
    out = []
    for l in data:
        if l[col] == key:
            out.append(l)
    return out
most_filter(0,data)
```

## Filter by least common in bit

```
def least_filter(col, data):
    def freq(line):
        value, freq = mode(line)
        if freq == len(line)/2: # if equal number of bits with both 1's and 0's
            return 1
        else:
            return value[0]
    '''Keep only the lines in data that have the LEAST common value in column "col"'''
    c = column(col, data)
    key = freq(c)
    if key == 0:
        key = 1
    else:
        key = 0
    out = []
    for l in data:
        if l[col] == key:
            out.append(l)
    return out
least_filter(0,data)
```

## Oxygen

```
def oxygen(data):
    for n in range(len(data[0])):
        data = most_filter(n, data)
        if len(data) == 1: # only one number left
            out = data[0]
            out = ''.join([str(q) for q in out])
            return int(out,2)
oxygen(data)
```

## CO2

```
def co2(data):
    for n in range(len(data[0])):
        data = least_filter(n, data)
        if len(data) == 1: # only one number left
            out = data[0]
            out = ''.join([str(q) for q in out])
            return int(out,2)
co2(data)
```

```
print('Answer part 2: {}'.format(oxygen(data)*co2(data)))
```
